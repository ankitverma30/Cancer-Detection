# Using PCA for transformation

```python
features=['concave points_worst','perimeter_worst','concave points_mean','radius_worst','perimeter_mean','area_worst','radius_mean','area_mean','concavity_mean','concavity_worst']
```

```python
x=data.loc[:,features].values
```

```python
data.diagnosis.head()
```

```python
y = data.loc[:,['diagnosis']].values
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

x = StandardScaler().fit_transform(x)
```

```python
pd.DataFrame(data = x, columns = features).head()
```

# PCA Projection to 4D

```python
pca = PCA(n_components=4)
```

```python
principalComponents = pca.fit_transform(x)
```

```python
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2','principal component 3','principal component 4'])
```

```python
principalDf.head()
```

```python
data.diagnosis.head()
```

```python
finaldata=pd.concat([principalDf,data.diagnosis],axis=1)
```

```python
finaldata.head()
```

```python
data['diagnosis'].value_counts()
```

# Visualiasing the PCA

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('4 Component PCA', fontsize = 20)


diagnosis = [0,1]
colors = ['r', 'g', 'b']
for diagnosis, color in zip(diagnosis,colors):
    indicesToKeep = finaldata['diagnosis'] == diagnosis
    ax.scatter(finaldata.loc[indicesToKeep, 'principal component 1']
               , finaldata.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
